```python
# Importar las librer√≠as necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Definir las rutas de los archivos CSV
file_paths = [
    'data_parte_2/2015.csv',
    'data_parte_2/2016.csv',
    'data_parte_2/2017.csv',
    'data_parte_2/2018.csv',
    'data_parte_2/2019.csv'
]

# Cargar los datos en un DataFrame unificado
df = pd.concat([pd.read_csv(file_path) for file_path in file_paths], ignore_index=True)

```

```python
#Inspeccionar los datos
print(df.head())
print(df.info())
print(df.describe())
```

```python
#Inspeccionar los tipo de datos
print(df.dtypes)
```

### Analizar la distribuci√≥n de los puntajes de felicidad

```python
sns.displot(df['Score'], kde=True)
plt.title('Distribuci√≥n de los puntajes de felicidad')
plt.show()
```

### An√°lisis de la puntuaci√≥n de felicidad promedio por regi√≥n

```python
df.groupby('Region')['Happiness Score'].mean().plot(kind='bar')
plt.title('Puntuaci√≥n de Felicidad Promedio por Regi√≥n')
plt.xlabel('Regi√≥n')
plt.ylabel('Puntuaci√≥n de Felicidad')
```

### An√°lisis de la esperanza de vida promedio por regi√≥n

```python
df.groupby('Region')['Health (Life Expectancy)'].mean().plot(kind='bar')
plt.title('Esperanza de Vida Promedio por Regi√≥n')
plt.xlabel('Regi√≥n')
plt.ylabel('Esperanza de Vida')
```

### Estudio de la percepci√≥n de corrupci√≥n promedio por regi√≥n

```python
df.groupby('Region')['Trust (Government Corruption)'].mean().plot(kind='bar')
plt.title('Percepci√≥n de Corrupci√≥n Promedio por Regi√≥n')
plt.xlabel('Regi√≥n')
plt.ylabel('Percepci√≥n de Corrupci√≥n')
```

### Dispersi√≥n entre Felicidad y Econom√≠a

```python
# Convertir la columna 'Happiness Score' a num√©rica
df['Happiness Score'] = pd.to_numeric(df['Happiness Score'])

# Gr√°fico de dispersi√≥n entre Felicidad y Econom√≠a
plt.figure(figsize=(10, 6))
plt.scatter(df['Economy (GDP per Capita)'], df['Happiness Score'])
plt.xlabel('Econom√≠a (GDP per C√°pita)')
plt.ylabel('Felicidad (Happiness Score)')
plt.title('Relaci√≥n entre Felicidad y Econom√≠a')
plt.show()
```

### Dispersi√≥n entre felicidad y salud

```python
# Convertir columnas a tipos de datos apropiados
df['Health (Life Expectancy)'] = df['Health (Life Expectancy)'].astype(float)

# Gr√°fico de dispersi√≥n
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Health (Life Expectancy)', y='Happiness Score', data=df)
plt.xlabel('Salud (Esperanza de Vida)')
plt.ylabel('Puntaje de Felicidad')
plt.title('Relaci√≥n entre Felicidad y Salud')
plt.show()
```

### Dispersi√≥n entre felicidad y confianza del gobierno

```python
# Convertir columnas a tipos de datos apropiados
df['Trust (Government Corruption)'] = df['Trust (Government Corruption)'].astype(float)

# Gr√°fico de dispersi√≥n
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Trust (Government Corruption)', y='Happiness Score', data=df)
plt.xlabel('Confianza en el Gobierno (Corrupci√≥n)')
plt.ylabel('Puntaje de Felicidad')
plt.title('Relaci√≥n entre Felicidad y Confianza en el Gobierno')
plt.show()
```
