```python
# Importar las librerías necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Definir las rutas de los archivos CSV
file_paths = [
    'data_parte_2/2015.csv',
    'data_parte_2/2016.csv',
    'data_parte_2/2017.csv',
    'data_parte_2/2018.csv',
    'data_parte_2/2019.csv'
]

# Cargar los datos en un DataFrame unificado
df = pd.concat([pd.read_csv(file_path) for file_path in file_paths], ignore_index=True)

```

```python
#Inspeccionar los datos
print(df.head())
print(df.info())
print(df.describe())
```

```python
#Inspeccionar los tipo de datos
print(df.dtypes)
```

### Analizar la distribución de los puntajes de felicidad

```python
sns.displot(df['Score'], kde=True)
plt.title('Distribución de los puntajes de felicidad')
plt.show()
```

### Análisis de la puntuación de felicidad promedio por región

```python
df.groupby('Region')['Happiness Score'].mean().plot(kind='bar')
plt.title('Puntuación de Felicidad Promedio por Región')
plt.xlabel('Región')
plt.ylabel('Puntuación de Felicidad')
```

### Análisis de la esperanza de vida promedio por región

```python
df.groupby('Region')['Health (Life Expectancy)'].mean().plot(kind='bar')
plt.title('Esperanza de Vida Promedio por Región')
plt.xlabel('Región')
plt.ylabel('Esperanza de Vida')
```

### Estudio de la percepción de corrupción promedio por región

```python
df.groupby('Region')['Trust (Government Corruption)'].mean().plot(kind='bar')
plt.title('Percepción de Corrupción Promedio por Región')
plt.xlabel('Región')
plt.ylabel('Percepción de Corrupción')
```

### Dispersión entre Felicidad y Economía

```python
# Convertir la columna 'Happiness Score' a numérica
df['Happiness Score'] = pd.to_numeric(df['Happiness Score'])

# Gráfico de dispersión entre Felicidad y Economía
plt.figure(figsize=(10, 6))
plt.scatter(df['Economy (GDP per Capita)'], df['Happiness Score'])
plt.xlabel('Economía (GDP per Cápita)')
plt.ylabel('Felicidad (Happiness Score)')
plt.title('Relación entre Felicidad y Economía')
plt.show()
```

### Dispersión entre felicidad y salud

```python
# Convertir columnas a tipos de datos apropiados
df['Health (Life Expectancy)'] = df['Health (Life Expectancy)'].astype(float)

# Gráfico de dispersión
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Health (Life Expectancy)', y='Happiness Score', data=df)
plt.xlabel('Salud (Esperanza de Vida)')
plt.ylabel('Puntaje de Felicidad')
plt.title('Relación entre Felicidad y Salud')
plt.show()
```

### Dispersión entre felicidad y confianza del gobierno

```python
# Convertir columnas a tipos de datos apropiados
df['Trust (Government Corruption)'] = df['Trust (Government Corruption)'].astype(float)

# Gráfico de dispersión
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Trust (Government Corruption)', y='Happiness Score', data=df)
plt.xlabel('Confianza en el Gobierno (Corrupción)')
plt.ylabel('Puntaje de Felicidad')
plt.title('Relación entre Felicidad y Confianza en el Gobierno')
plt.show()
```
